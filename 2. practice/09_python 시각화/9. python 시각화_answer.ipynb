```python
# 이미지 크기 변경
plt.figure(figsize=(10,8))

# plot 그리기
plt.plot(["Seoul","Paris","Seattle"], [10,35,20], '-',color='orange')
plt.plot(["Seoul","Paris","Seattle"], [30,25,55], '--',color='blue')
plt.plot(["Seoul","Paris","Seattle"], [20,10,40], '-.',color='red')
plt.plot(["Seoul","Paris","Seattle"], [5,24,10], ':',color='magenta')

# x축 이름 지정
plt.xlabel('City')

# y축 이름 지정
plt.ylabel('Response')

# 제목 지정
plt.title('Experiment Result')

# 범례 추가
# upper, center, lower / left, center, right
plt.legend(['a', 'b', 'c', 'd'], loc='upper left')

# 축값 변경
plt.ylim(0, 100)

# grid 표사
plt.grid(True)

plt.show()
```

```python
# 이미지 크기 변경
plt.figure(figsize=(10,8))

# plot 그리기
plt.plot(["21st","22nd","23th"], [x for x in range(3)], 'c^--')
plt.plot(["21st","22nd","23th"], [x for x in range(3, 6)], 'ys-.')
plt.plot(["21st","22nd","23th"], [x for x in range(6, 3, -1)], 'ro-')
plt.plot(["21st","22nd","23th"], [x for x in range(3, 0, -1)], 'g<:')

# x축 이름 지정
plt.xlabel('index')

# y축 이름 지정
plt.ylabel('value')

# 제목 지정
plt.title('Result')

# 범례 추가
# upper, center, lower / left, center, right
plt.legend(['0to2', '3to5', '6to4', '3to1'], loc='upper right', fontsize=15)

# 축값 변경
plt.ylim(0, 8)

# grid 표사
plt.grid(True)

plt.show()
```

## 실습

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
```

```python
my_colors = ['blue' if x>5 else 'red' for x in iris.sepal_length[:20]]
iris.sepal_length[:20].plot(kind='bar', rot=0, color=my_colors)
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("sepal_length")
plt.show()
```

## 2. 왕좌의 게임 데이터 분석 및 시각화 실습
### 1) 데이터 불러오기

```python
battles = pd.read_csv('data/data-society-game-of-thrones/battles.csv')
death = pd.read_csv('data/data-society-game-of-thrones/character-deaths.csv')
```

### 2) 책의 챕터(Book of Death)별로 사망자 수 추이 시각화

```python
book_of_death = death['Book of Death'].value_counts().sort_index()
book_of_death
```

```python
axes1 = book_of_death.plot()
axes1.set_xlim(0, 6)
axes1.set_xlabel('Book Number')
axes1.set_ylabel('number of death')
axes1.set_title('number of death for book number')
```

### 3) 대규모 전쟁의  공격진영과 수비 진영의 숫자 시각화
#### 대규모 전쟁 = 공격과 수비 모두 합쳐서 10,000명 이상 참가한 전쟁

```python
# 1. 대규모 전쟁만 선택
big_battles = battles[battles['attacker_size'] + battles['defender_size'] >= 10000]
```

```python
# 2. 필요한 컬럼만 선택
big_battles = big_battles[['name', 'attacker_size', 'defender_size']]
```

```python
big_battles
```

```python
# big_battles 시각화
big_battles.set_index('name', inplace = True)
```

```python
big_battles.plot(kind='barh', stacked = True)
```

### 4) (실습) 가문 별로 전쟁을 참여한 횟수 시각화

```python
a1 = battles.pivot_table(index='attacker_1', aggfunc = 'count', values = 'name')
a2 = battles.pivot_table(index='attacker_2', aggfunc = 'count', values = 'name')
a3 = battles.pivot_table(index='attacker_3', aggfunc = 'count', values = 'name')
a4 = battles.pivot_table(index='attacker_4', aggfunc = 'count', values = 'name')

d1 = battles.pivot_table(index='defender_1', aggfunc = 'count', values = 'name')
d2 = battles.pivot_table(index='defender_2', aggfunc = 'count', values = 'name')
d3 = battles.pivot_table(index='defender_3', aggfunc = 'count', values = 'name')
d4 = battles.pivot_table(index='defender_4', aggfunc = 'count', values = 'name')

new = pd.concat([a1, a2, a3, a4, d1, d2, d3, d4], axis=1, join='outer')
new = new.fillna(0)
new
final = DataFrame(new.apply(sum, axis = 1))
final
```

```python
final.plot(kind = 'bar')
```

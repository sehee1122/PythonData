```python
# 실습 데이터 적재
data = pd.read_excel('data/인구수예제.xlsx')
data.head()
```

## 1. groupby
## (dataframe).groupby('집단별로 묶고 싶은 기준 column')['이후에 연산을 하고 싶은 column'].함수

```python
# 연도별 총인구의 합
# 집단별로 묶고 싶은 기준 : '연도'
# 연산을 하고 싶은 column : '총인구'
# 사용할 함수 : 'sum()'
data.groupby('연도')['총인구'].sum()
```

```python
pd.DataFrame(data.groupby('연도')['총인구'].sum())
```

```python
# reset_index() : index의 값을 column으로 이동
pd.DataFrame(data.groupby('연도')['총인구'].sum()).reset_index()
```

```python
# 자치구별 총인구의 평균
data.groupby('자치구')['총인구'].mean()
```

## 2. pivot_table
## (dataframe).pivot_table(index='', aggfunc='', values='')
### index='집단별로 묶고 싶은 기준 column', 
### aggfunc=함수, 
### values='이후에 연산을 하고 싶은 column'

```python
# 연도별 총인구의 합
data.pivot_table(index = '연도', aggfunc = 'sum', values = '총인구')
```

```python
# 자치구별 총인구의 평균
data.pivot_table(index = '자치구', aggfunc = 'mean', values = '총인구')
```

## group, pivot_table 예제

## 1) 연도별 전체의 남자인구, 여자인구, 총인구 수 합 구하기

```python
# groupby
data.groupby('연도')[['남자인구', '여자인구', '총인구']].sum()
```

```python
# pivot_table
data.pivot_table(index = '연도', aggfunc = 'sum', values = ['남자인구', '여자인구', '총인구'])
```

## 2) 도시/자치구별 평균 총인구수

```python
# groupby
data.groupby(['도시', '자치구'])['총인구'].mean()
```

```python
# pivot_table
data.pivot_table(index = ['도시', '자치구'], aggfunc = 'mean', values = ['총인구'])
```

## 3)  도시별, 연도별로 총인구수 합 출력

```python
# groupby
pd.DataFrame(data.groupby(['도시', '연도'])['총인구'].sum()).unstack('도시')
```

```python
# pivot_table
data.pivot_table(index = '연도', columns = '도시', aggfunc = 'sum', values = '총인구')
```

# 실습

## 1. 자치구별로 평균 총 인구수

```python
# groupby 사용
data.groupby(['자치구'])['총인구'].mean()
```

```python
# pivot_table 사용
data.pivot_table(index = '자치구', aggfunc = 'mean', values = '총인구')
```

## 2. 도시/자치구별 평균 남자인구와 여자인구수 구하기

```python
# groupby 사용
data.groupby(['도시', '자치구'])['남자인구', '여자인구'].mean()
```

```python
# pivot_table 사용
data.pivot_table(index = ['도시', '자치구'], aggfunc = 'mean', values = ['남자인구', '여자인구'])
```

### 3. 도시/자치구 별로 남자인구의 평균을 구한 후, 남자인구가 가장 많은 도시/자치구 5개를 출력하시오. 

```python
# groupby 사용
data.groupby(['도시', '자치구'])['남자인구'].mean().sort_values(ascending = False).head(5)
```

```python
# pivot_table 사용
data.pivot_table(index = ['도시', '자치구'], aggfunc = 'mean', values = '남자인구').sort_values(by = '남자인구', ascending = False).head(5)
```

## 함수를 여러개 적용하고 싶은 경우

```python
# 연도, 도시별로 총인구의 평균과, 표준편차 구해보기
```

```python
# pivot_table
# mean: 평균 / std: 표준편차

data.pivot_table(index = ['연도', '도시'], aggfunc = ['mean', 'std'],
                values = '총인구')
```

```python
data.pivot_table(index = '연도', columns = '도시', aggfunc = ['mean', 'std'],
                values = '총인구')
```

```python
# groupby 함수에서는 agg() 함수를 적용하고 안에 list형식으로 넣어준다

data.groupby(['연도', '도시'])['총인구'].agg(['mean', 'std']).unstack(level = 1)
```

```python
# 실습. 도시/자치구별 기준으로 남녀차이의 평균이 가장 많이 나는 도시/자치구 3개를 찾아보기
# Hint: abs() = 절대값을 구하는 함수

data['남녀차이'] = abs(data['남자인구'] - data['여자인구'])
```

```python
data.groupby(['도시', '자치구'])['남녀차이'].mean().sort_values(ascending = False).head(3)
```

```python
temp = data.pivot_table(index = ['도시', '자치구'], aggfunc = 'mean', values = '남녀차이')
temp.sort_values(by = '남녀차이', ascending = False).head(3)
```

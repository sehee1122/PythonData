## 아래와 같은 결과를 만들어주세요

```python
# 아래 셀에서 작성해주세요
NC_final
```

```python
NC = pd.read_excel('data/NC Dinos.xlsx',sheet_name=None)
NC13,NC14,NC15=NC.values()
NC13.set_index('선수명',inplace=True)
NC14.set_index('선수명',inplace=True)
NC15.set_index('선수명',inplace=True)

NC_hit = pd.concat([NC13['안타'], NC14['안타'], NC15['안타']], axis=1, sort=True).fillna('-')
NC_hit.columns = [['안타', '안타', '안타'], [2013, 2014, 2015]]
NC_hit.columns.set_names(['','연도'], inplace=True)
NC_hit.index.set_names(['선수명'], inplace=True)
NC_hit

NC_homerun = pd.concat([NC13['홈런'], NC14['홈런'], NC15['홈런']], axis=1, sort=True).fillna('-')
NC_homerun.columns = [['홈런', '홈런', '홈런'], [2013, 2014, 2015]]
NC_homerun.columns.set_names(['','연도'], inplace=True)
NC_homerun.index.set_names(['선수명'], inplace=True)
NC_homerun

NC_final = pd.concat([NC_hit, NC_homerun], axis=1).fillna('-')
NC_final
```

## 1학기 때 언어, 수리 평균은 문과/이과 중 누가 더 높을까?
### HINT: 처음부터 이과이면서 1학기 성적을 추출하려 하지말고 이과를 출력한 다음에 과목 성적 추출

```python
lee=python_study.xs('이과',axis=0,level=1)
lee_1=lee.xs('1학기',axis=1,level=0)
lee_kor=lee_1['언어'].mean()
lee_math=lee_1['수리'].mean()

mun=python_study.xs('문과',axis=0,level=1)
mun_1=mun.xs('1학기',axis=1,level=0)
mun_kor=mun_1['언어'].mean()
mun_math=mun_1['수리'].mean()

print("이과 언어 평균 :",lee_kor,"문과 언어 평균 :",mun_kor)
print("이과 수리 평균 :",lee_math,"문과 수리 평균 :",mun_math)
```

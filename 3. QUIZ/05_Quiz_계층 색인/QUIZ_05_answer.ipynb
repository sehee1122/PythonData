```python
# 데이터 불러오기
data = pd.read_excel('data/NC Dinos.xlsx', sheet_name=None)
NC13, NC14, NC15 = data.values()

# 필요없는 컬럼 제거
NC13.drop('Unnamed: 0', axis=1, inplace=True)
NC14.drop('Unnamed: 0', axis=1, inplace=True)
NC15.drop('Unnamed: 0', axis=1, inplace=True)

# 연도 추가
NC13['연도'] = 2013
NC14['연도'] = 2014
NC15['연도'] = 2015

NC13.set_index('선수명', inplace=True)
NC14.set_index('선수명', inplace=True)
NC15.set_index('선수명', inplace=True)
```

## 1학기 때 언어, 수리 평균은 문과/이과 중 누가 더 높을까?
### HINT: 처음부터 이과이면서 1학기 성적을 추출하려 하지말고 이과를 출력한 다음에 과목 성적 추출

```python
python_study.xs('문과', axis=0, level=1)[('1학기', '언어')].mean()
```

```python
def check(subject):
    society = python_study.xs('문과', axis=0, level=1)[('1학기', subject)].mean()
    science = python_study.xs('이과', axis=0, level=1)[('1학기', subject)].mean()
    
    if society > science:
        print('%s 과목은 문과의 성적이 이과보다 높습니다' % subject)
    elif society < science:
        print('%s 과목은 이과의 성적이 문과보다 높습니다' % subject)
    else:
        print('%s 과목은 문과의 성적과 이과의 성적은 같습니다' % subject)
```

```python
check('언어')
```

```python
check('수리')
```

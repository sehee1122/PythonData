```python
data.columns = ['city', 'province', 'year', 'male', 'female','total']
data = data.replace(['서울', '부산'], ['Seoul', 'Busan'])
data.head()
```

## 2. Seoul과 Busan의 2013~2017년도의 총인구 합 선그래프를 그려주세요

```python
# 아래 셀에서 작성해주세요
```

```python
df = pd.DataFrame(data.groupby(['city', 'year'])['total'].sum())
df
```

```python
df = pd.DataFrame(data.groupby(['city', 'year'])['total'].sum().unstack('city'))
df
```

```python
axes = df.plot()

axes.set_xlabel('year')
axes.set_ylabel('population')
axes.set_title('Seoul/Busan population')
axes.legend(['Busan', 'Seoul'], loc = "center right", title = "city")

axes
```

```python
baseball = pd.read_excel("data/NC Dinos.xlsx", sheet_name = "2013")
baseball
```

```python
def draw(num, col, x, y):
    ax = fig.add_subplot(2, 2, num)
    
    ax.scatter(baseball[x], baseball[y], color = col)
```

```python
fig = plt.figure(figsize=(8, 8))

draw(1, 'red', '득점', '타점')
draw(2, 'blue', '타수', '홈런')
draw(3, 'yellow', '홈런', '도루')
draw(4, 'pink', '볼넷', '삼진')
```

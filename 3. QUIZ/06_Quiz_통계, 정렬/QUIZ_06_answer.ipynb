## 10. 7번의 결과 데이터프레임을 활용해서 province를 입력하면 contact_number가 가장 높은 5개의 city를 출력해주는 함수를 만들어주세요

```python
# 아래 셀에서 작성해주세요
print_city('Seoul')
```

```python
# 아래 셀에서 작성해주세요
print_city('Busan')
```

```python
def print_city(province):
    temp = patient_group.xs(province, axis=0, level=0).sort_values(by='contact_number', ascending=False).head(5).index
    for n, v in enumerate(temp):
        print('%d 번째 : %s ' % (n+1, v))
```

## 11. patientInfo와 PatientRoute를 patient_id를 기준으로 inner join 해주세요

## 12. 위에서 merge한 파일에서 각 column마다 NaN의 개수와 비율을 구하기 위해 아래와 같은 data frame을 만들어주세요

```python
# 아래 셀에서 작성해주세요
na
```

```python
na = {'index':[x for x in range(patient_merge.shape[1])],
      'column type':patient_merge.dtypes,
      'null values(num)':patient_merge.isnull().sum(),
      'null values(%)':patient_merge.isnull().sum() * 100 / patient_merge.shape[0]}

na = pd.DataFrame(na).T
```

## 13. null values(%)가 100인 column은 제거해주세요

```python
null_100 = na.loc['null values(%)']
drop_col = null_100[null_100 == 100].index
patient_merge.drop(drop_col, axis=1, inplace=True)
```
